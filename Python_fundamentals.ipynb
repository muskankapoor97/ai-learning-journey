```python
numbers = [4, 7, 2, 9, 1, 5]
numb =[]
def get_average(nums):
  total = 0
  if nums == []:
    return None
  for n in nums:
    total = total + n
  return total/len(nums)
result = get_average(numb)
print("Average:", result)
```

```python
def normalize(nums):
  if not nums:
    return []
  minimum = min(nums)
  maximum = max(nums)
  if minimum == maximum:
    return nums
  return [((num - minimum) / (maximum - minimum)) for num in nums]

print(normalize([10, 20, 30, 40, 50]))
print(normalize([]))
print(normalize([5, 5, 5]))
```

```python
def standardize(nums):
  if not nums:
    return []
  mean = sum(nums) / len(nums)
  std_dev = (sum((x - mean) ** 2 for x in nums) / len(nums)) ** 0.5
  if std_dev == 0:
    return nums
  return [(x - mean) / std_dev for x in nums]

print(standardize([10, 20, 30, 40, 50]))
print(standardize([]))
print(standardize([5, 5, 5]))
```
